```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, expon
```

```python
customers_per_hour = 15
hours = 1000

timestamp = np.random.uniform(low=0, high=60*hours, size=customers_per_hour*hours)
timestamp = sorted(timestamp)
tdf = pd.DataFrame({'timestamp':timestamp})
tdf['hour'] = tdf['timestamp']//60
hour_count = tdf.groupby(by='hour')['hour'].count()
freq = hour_count.value_counts()
prob = freq/sum(freq)

plt.figure(figsize=(4,3))
plt.bar(prob.index, prob)

xs = np.arange(0,30)
ys = [poisson.pmf(x,customers_per_hour) for x in xs]

plt.plot(xs, ys, color='C1', marker='o')
plt.show()


interval = []
for i in range(len(timestamp) -1):
	interval.append(timestamp[i+1]-timestamp[i])
	
plt.figure(figsize=(4,3))
n, bins, patches = plt.hist(interval, density=True, bins=30, edgecolor='k')

plt.show()

expon.pdf(20, scale= 60/customers_per_hour)
```

```python
# Randomly sample 30 surveys
df = pd.read_csv('survey.csv')
print(df['activity_level'].mean())

sample_means = []
for i in range(10000):
	sample_means.append(df.sample(30)['activity_level'].mean())


np.mean(sample_means)

plt.figure(figsize=(4,3))
plt.hist(sample_means, bins= 30, edgecolor='k')
plt.show()
```
